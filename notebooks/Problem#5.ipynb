Import the following four files containing UV-vis spectra of four food dyes with the first column listing the wavelengths (nm) and the second column containing the absorbances. Each file contains data from 400-850 nm in 1 nm increments.

'red40.csv', 'green3.csv', 'blue1.csv', 'yellow6.csv'


a) Concatenate the files into a single DataFrame with the first column as the wavelength (nm) and the other four columns as the absorbances for each dye.

b) Replace the column headers with meaningful labels.

```python
import pandas as pd
```

```python
path = 'e:/Python/Python For Chemistry/data'
red40 = pd.read_csv(f'{path}/red40.csv')
green3 = pd.read_csv(f'{path}/green3.csv')
blue1 = pd.read_csv(f'{path}/blue1.csv')
yellow6 = pd.read_csv(f'{path}/yellow6.csv')
```

```python
red40.columns = ['Wavelength_nm', 'Absorbance']
yellow6.columns = ['Wavelength_nm', 'Absorbance']
green3.columns = ['Wavelength_nm', 'Absorbance']
blue1.columns = ['Wavelength_nm', 'Absorbance']
```

Converting into data frames so to merge them

```python
df_red40 = pd.DataFrame(red40)
df_green3 = pd.DataFrame(green3)
df_blue1 = pd.DataFrame(blue1)
df_yellow6 = pd.DataFrame(yellow6)
```

Merging Datasets.
1 wavelength vs absorbance

1. renaming according to dyes

```python
red40   = red40.rename(columns={'Absorbance': 'Red40'})
green3 = green3.rename(columns={'Absorbance': 'Green3'})
blue1  = blue1.rename(columns={'Absorbance': 'Blue1'})
yellow6 = yellow6.rename(columns={'Absorbance': 'Yellow6'})
```

2. setting wavelength as index

```python
red40 = red40.set_index('Wavelength_nm')
green3 = green3.set_index('Wavelength_nm')
blue1 = blue1.set_index('Wavelength_nm')
yellow6 = yellow6.set_index('Wavelength_nm')
```

2. Concatenating

```python
uvvis_all = pd.concat(
    [red40, green3, blue1, yellow6],
    axis=1
)
```

```python
print(uvvis_all.to_string())
```
